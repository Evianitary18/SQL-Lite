**No. 1**

1. Sebutkan 5 kategori yang memiliki rata-rata jumlah potongan harga terbesar. Sebutkan juga masing-masing nilai dari kategori tersebut!
Expected output pada query (yang dimaksud dengan expected output adalah nama setiap kolom yang harus ada pada hasil query)
a. category
b. avg_discount

```python
# Langkah pertama kita harus membuat database dan mengimpor data terlebih dahulu.

import pandas as pd
import sqlite3
```

```python
# Langkah kedua kita harus membaca file excelnya
nama_file = 'amazon_sales.xlsx'
df = pd.read_excel(nama_file, sheet_name = 'amazon')
```

```python
# Langkah ketiga, lihat dulu file yang ada di dalam amazon_sales.xsls
df.head()
```

```python
# Langkah keempat, membuat koneksi
conn = sqlite3.connect('amazon_sales.db')
df.to_sql('amazon_sales', conn, if_exists = 'replace', index = False)
```

```python
# Langkah kelima mencari 5 kategori dengan rata-rata potongan harga terbesar
query1= """
SELECT category, AVG(discount_percentage) AS avg_discount
FROM amazon_sales
GROUP BY category
ORDER BY avg_discount DESC
LIMIT 5;
"""
```

```python
# Langka keenam ialah menampilkan hasil dari query
hasil = pd.read_sql_query(query1, conn)
print(hasil)
```

```python
pd.read_sql_query(query1, conn)
```

**No. 2**
Berapa jumlah user yang memberikan review yang memuat kata 'Good Product' pada produk dengan kategori Computers & Accessories dan Office Products ?
Expected output pada query :
- category
- jumlah_review

```python
query2 = """
SELECT
  category,
  COUNT(DISTINCT review_content) AS jumlah_review
FROM
  amazon_sales
WHERE
  (category = 'Computers & Accessories' OR category = 'Office Products')
AND review_content LIKE '%Good Product%'
GROUP BY
  category;
"""
```

```python
pd.read_sql_query(query2, conn)
```

**No. 3**
Tentukan 5 bulan yang memiliki total sales/revenue terendah pada tahun 2004 (dari seluruh kategori produk kecuali Office Products dan Home & Kitchen). Sebutkan juga nilai setiap revenue nya!
Expected output pada query
- total_sales
- bulan
- tahun

```python
query3 = """
SELECT
  strftime('%Y', order_date) AS tahun,
  strftime('%m', order_date) AS bulan,
  SUM(discounted_price * qty) AS total_sales
FROM
  amazon_sales
WHERE
  strftime('%Y', order_date) = '2004'
AND
  category NOT IN ('Office Products', 'Home & Kitchen')
GROUP BY
  order_date
ORDER BY
    total_sales ASC
LIMIT 5;
"""
```

```python
pd.read_sql_query(query3, conn)
```

**No. 4**
Sebutkan kategori apa yang rata-rata waktu pengirimannya paling cepat dan lambat (Tanpa kategori Car & Motorbike dan Toys & Games).
Tips : Rata-rata waktu pengiriman dapat dibulatkan tanpa ada angka di belakang koma

Expected output pada query
- category
- arrival_date
- order_date
- lama_pengiriman

```python
# Waktu pengiriman paling cepat
query4 = """
SELECT
  category,
  ROUND(AVG(strftime('%d', arrival_date) - strftime('%d', order_date))) AS lama_pengiriman
FROM
  amazon_sales
WHERE
  category NOT IN ('Car & Motorbike', 'Toys & Games')
GROUP BY
  category
ORDER BY
  lama_pengiriman ASC;
"""
```

```python
pd.read_sql_query(query4, conn)
```

```python
# pengiriman paling lambat

query0 = """
SELECT
  category,
  ROUND(AVG(CAST(strftime('%s', arrival_date) AS INTEGER) - CAST(strftime('%s', order_date) AS INTEGER)) / 86400) AS lama_pengiriman
FROM
  amazon_sales
WHERE
  category NOT IN ('Car & Motorbike', 'Toys & Games')
GROUP BY
  category
ORDER BY
  lama_pengiriman DESC;
"""
```

```python
pd.read_sql_query(query0, conn)
```

**No. 5**
Hitunglah AOV pada setiap tahunnya, namun data yang digunakan haruslah data dari awal hingga pertengahan bulan (tanggal 01 - 15) untuk setiap bulannya

Expected output pada query
- total_sales
- total_order
- aov
- year

```python
# Berikut merupakan query untuk menampilkan data sesuai expected output.
query5 = """
SELECT
  strftime('%Y', order_date) AS year,
  SUM(qty) AS total_sales,
  AVG(discounted_price * qty) AS AOV,
  SUM(discounted_price * qty) AS total_order
FROM
  amazon_sales
WHERE
  strftime('%d', order_date) BETWEEN '01' AND '15'
GROUP BY
  year
ORDER BY
  year;
"""
```

```python
pd.read_sql_query(query5, conn)
```
