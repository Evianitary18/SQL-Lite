Macam - macam join
1. Inner Join
2. Left outer join
3. Left outer join without intersection
4. Right outer join
5. Right outer join without interscetion

```python
# 1. Inner join

SELECT nama_kolom
from table1
inner join table2
on table1.nama_kolom = table2.nama_kolom
```

```python
pd.read_sql_query("SELECT i.Customerid,c.FirstName|| ' ' ||c.Lastname as FullName, Invoiceid, InvoiceDate, Total from invoices as i inner join as c o on i.CustomerId=c.CustomerId", con)
```

```python
#2. Left outer join yang akan menampilkan gabungan dari seluruh baris data dari tabel sebelah kiri/table pertama memenuhi kondisi on dan data dari sebelah kanan memenuhi kondisi join
SELECT nama_kolom
FROM tabelA
Left outer join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
```

```python
# Contoh kita aakan menggunakan kolom artis dan album dengan identifier kolom artis id
# Jadi album dialiaskan x dan artis dialiaskan y
pd.read_sql_query("SELECT * from(select * from album limit 10) x left outer join(select * form artists limit 5) y using (ArtistId)limit 10", con)
```

```python
# 3. Left outer join without intersection
# Artinya hanya menampilkan baris data dari tabel kiri atau tabel pertama yang memenuhi kondisi on dan join, tanpa data dari tabel kanan yang memenuhi kondisi join
SELECT nama_kolom
FROM tabelA
Left outer join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
WHERE tabelB.nama_kolom is null

#bedanya ditabel kanan ditambahi null karena nilai kolom kanan null
```

```python
pd.read_sql_query("SELECT * from(select * from album limit 10) x left outer join(select * form artists limit 5) y using (ArtistId) where Name is null", con)
```

```python
# 4. Right Outer Join
# Menampilkan gabungan dari seluruh baris data dari tabel sebelah kanan atau tabel kedua yang memenuhi kondisi on, dan data dari tabel kiri atau pertama memenuhi kondisi join
SELECT nama_kolom
FROM tabelA
Right outer join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
```

```python
# 5. Right Outer Join Without Intersectiom
# Hanya akan menampilkan baris data dari tabel kanan atau tabel kedua yang memenuhi kondisi on dan kondisi join. Tanpa data dari tabel kiri yang memenuhi kondisi join
SELECT nama_kolom
FROM tabelA
Right outer join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
WHERE tabelA.nama_kolom is null
```

```python
# SQLITE nggak mendukung right outer join without section
```

```python
# 6. Full outer join
# Menampilkan seluruh baris data dari kedua tabel yang memenuhoi kondisi on
SELECT nama_kolom
FROM tabelA
Full join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
WHERE tabelA.nama_kolom is null no tabelB.nama_kolom is null
```

```python
# SQLITE nggak mendukung full outer join.
```

```python
# 7.Full Outer join witout intersection
# kedua tabel memenuhi kondisi on tanpa data yang memiliki nilai null

SELECT nama_kolom
FROM tabelA
Full join tabelB
on tabelA.nama_kolom = tabelB.nama_kolom
WHERE tabelA.nama_kolom is null or tabelB.nama_kolom is null
```

Macam -macam join lainnya

```python
# Cross join
# Akan membandingkan setipa baris pada tabel kiri atau tabel pertama dengan setiap baris pada tabel kanan atau tabel kedua
```

cross join.png

```python
# Sintaks
SELECT nama_kolom
FROM tabelA
Cross join tabelB;
```

```python
# contoh cross join
# Database yang digunakan masih chinook.db
pd.read_sql_query("SELECT * from(select distinct Title from album) cross join (select * from media_types) limit 10", con)
```

```python
# Self Join
# Menggabung kan dengan tabel itu sendiri
# Seolah-olah menggabungkan dua salinan yang sama, tapi kita sebenernya gak buat salinan

SELECT T1.nama_kolom, T2.nama_kolom,
FROM tabelA T1, tabelA T2
WHERE T1.nama_kolom = T2.nama_kolom;
```

```python
# Misal mau bikin tim yang terdiri dari 2 orang dengan data dari tabel employees

pd.read_sql_query("SELECT E1.FirstName, E2.FirstName from employees E1, employees E2 where E1.EmployeeId!=E2.EmployeeId limit 10", con)
```
