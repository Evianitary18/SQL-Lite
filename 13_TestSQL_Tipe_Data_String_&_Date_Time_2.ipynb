Fungsi untuk tipe data datetime:
1. DATE()
2. TIME()
3. DATETIME()
4. STRFTIME()

1. DATE()
Memperoleh nilai tanggal dari suatu data yang ditampilkan dalam format "YYYY-MM-DD" dengan Y adalah tahun, M adalah bulan dan D adalah Day

```python
# Misal ingin tahu sekarang tanggal berapa
pd.read_sql_query("SELECT DATE('now')", con)
```

```python
# Jika ingin mendapatkan data dari suatu waktu
pd.read_sql_query("SELECT DATE('2021-09-01 09:34:21')", con)
```

```python
# Untuk memperoleh tanggal dari kolom invoices date
pd.read_sql_query("SELECT CustomerId, InvoiceDate, date(InvoiceDate)as Date from invoices", con)
```

2. TIME
Memperoleh nilai waktu dari suatu data yang ditampilkan dalam format "HH:MM:SS" dengan "H" adalah "Hour", "M" adalah "Minute", dan "S" adalah Second.

```python
# Jika sekarang ingin tahu waktu
pd.read_sql_query("SELECT TIME('now')", con)
```

```python
pd.read_sql_query("SELECT TIME('2017-08-15 19:30:10)", con)
```

```python
# Mengambil nilai waktu kapan employee di Hire berdasarkan kolom date
pd.read_sql_query("SELECT EmployeeId, HireDate, time(HireDate) as Time from employees", con)
```

```python
pd.read_sql_query("SELECT HireDAte, TIME(HireDate) as HireTime from employee", con)
```

3. DATETIME
Digunakan untuk memperoleh nilai tanggal dan waktu dari suatu data yang ditampilkan dalam format "YYYY-MM-DD HH:MM:SS"

```python
# Misalkan kita ingin mengatahui tanggal dan waktu dari Invoice dat saat ini
# +7 hour karena menyesuaikan WIB
# Fungsi date time juga anyak digunakan untuk mengkonversi data tanggal dan waktu yang tadinya bertipe string menjadi datetime.
pd.read_sql_query("SELECT InvoiceDAte, DATETIME('now') as Today_Time UTC, DATETIME('now', '+7 hour') as Today_Time from invoices", con)
```

4. STRFTIME
Memperoleh nilai tanggal dan atau waktu yang formatnya disesuaikan dengan format string yang telah ditentukan dalam argumen pertama

iiiiiiiiiiiiiii.png

```python
# Menghitung usia karyawan berdasarkan data tanggal lair mereka
pd.read_sql_query("SELECT Employee, FirstName, LastName, BirthDate, STRFTIME('%Y', 'now')" - STRFTIME('%Y', 'HireDate') as Age from employee", con)
```
