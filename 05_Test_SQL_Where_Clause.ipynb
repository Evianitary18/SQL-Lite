Apa fungsi klausa WHERE?

> Where digunakan untuk melakukan filter terhadap data yang ditampilkan dengan menyatakan kondisi atau syarat tertentu.



```python
# SINTAKS
SELECT nama_kolom FROM nama_tabel WHERE kondisi;
```

Kita bisa menggunakan berbagai macam operator WHERE mulai dari

"=", "<", ">", lebih dari sampai dengan kurang dari sama dengan, "<>"/"!="

**ATURAN**

- Jika ingin menggunakan tanda "=" untuk membandingkan nilai yang berupa string atau karakter, jangan lupa berikan tanda petik pada nilai-nilai yang dibandingkan.
Contoh = UserName = 'Ayu'

- Tanda"<>" atau "!=" digunakan jika kita tidak ingin memilih data yang sama.

- Tanda "<" dan ">" digunakan untuk data bertipe numerik (angka).
